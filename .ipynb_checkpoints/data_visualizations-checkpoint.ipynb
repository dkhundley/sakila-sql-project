# Hundley - Sakila SQL Visualizations
This notebook is taking the CSV results of the SQL files to create visualizations for our final presentation slides.

## Initial Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

## Visualization 1

```python
# Importing data
df1 = pd.read_csv('data/question1.csv')
```

```python
# Gathering rental counts per genre
df1 = df1.groupby(['genres']).sum().reset_index()
```

```python
# Visualizing data in matplotlib
plt.bar(x = df1['genres'], height = df1['rental_count']);
plt.title('Total Rentals by Family-Oriented Genres');
plt.ylabel('Total Rentals');
plt.xlabel('Family-Oriented Genre');
plt.ylim(bottom = 800);
```

## Visualization 2

```python
# Importing data
df2 = pd.read_csv('data/question2.csv')
```

```python
# Gathering number of movies in each quartile
df2 = df2.groupby(['standard_quartile']).sum().reset_index()
```

```python
# Visualizing data in matplotlib
plt.bar(x = df2['standard_quartile'], height = df2['rental_duration']);
plt.title('Quartiled Rentals by Family-Oriented Genres');
plt.ylabel('Number of Rentals in Quartile');
plt.xlabel('Quartiles');
```

## Visualization 3

```python
# Importing data
df3 = pd.read_csv('data/question3.csv')
```

```python
# Visualzing data in Seaborn
ax = sb.barplot(data = df3, x = 'genre', y = 'count', hue = 'standard_quartile');
ax.set_title('Number of Family-Oriented Movies Per Quartile');
ax.set_xlabel('Genres');
ax.set_ylabel('Number of Movies in Quartile')
plt.legend(title = 'Quantile', loc = 'best');
```

## Visualization 4

```python
# Importing data
df4 = pd.read_csv('data/question4.csv')
```

```python
# Visualizing data using seaborn
ax = sb.barplot(data = df4, x = 'rental_month', y = 'count_rentals', hue = 'store_id', order = [5, 6, 7, 8, 2])
ax.set_title('Number of Store Rentals Each Month');
ax.set_xlabel('Month / Year');
ax.set_ylabel('Number of Rentals')
ax.set_xticklabels(labels = ['May 2005', 'June 2005', 'July 2005', 'August 2005', 'February 2006'], rotation = 45);
plt.legend(title = 'Store ID', loc = 'best');
```
