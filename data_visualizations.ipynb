This notebook is taking the CSV results of the SQL files to create visualizations for our final presentation slides.

## Initial Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Visualization 1

```python
# Importing data
df1 = pd.read_csv('data/question1.csv')
```

```python
# Gathering rental counts per genre
df1 = df1.groupby(['genres']).sum().reset_index()
```

```python
# Visualizing data in seaborn
plt.bar(x = df1['genres'], height = df1['rental_count']);
plt.title('Total Rentals by Family-Oriented Genres');
plt.ylabel('Total Rentals');
plt.xlabel('Family-Oriented Genre');
plt.ylim(bottom = 800);
```
